```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data/ECGu.txt')  # Load the ECG data from the text file
print(data) # Print the ECG data
print(data.shape) # Print the number of rows and columns
```

```python
# conversion of data from A/U to mV (1 A/U = 1.024 mV)
data = data * 1024 
```

```python
# create a time vector t at the same length as ECG and with sampling rate of 1000 HZ
FS = 1000  # Hz
t = np.arange(0,data.shape[0]) / FS
print(t)
```

```python
# plot the ECG  data as a function of time
plt.plot(t, data)
plt.xlabel('Time (s)')
plt.ylabel('ECG (mV)')
plt.title('ECG Signal')
plt.grid()
plt.show()
```
